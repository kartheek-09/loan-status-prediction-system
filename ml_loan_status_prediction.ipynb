```python
# we want to convert ccategorical columns to numerical values
ldf.replace({"Married":{"No":0,"Yes":1},"Gender":{"Male":1,"Female":0,},"Self_Employed":{"No":0,"Yes":1},"Property_Area":{"Rural":0,"Semiurban":1,"Urban":2},"Education":{"Graduate":1,"Not Graduate":0}},inplace=True)
```

```python
ldf
```

```python
x=ldf.drop(columns=["Loan_ID","Loan_Status"],axis=1)
y=ldf["Loan_Status"]
```

```python
print(x)
```

```python
print(y)
```

```python
# splitting theee data set into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
classifier=svm.SVC(kernel="linear")
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on training data:",training_data_accuracy)
```

```python
x_test_prediction=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on testing data:",test_data_accuracy)
```
